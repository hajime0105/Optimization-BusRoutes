```python
import numpy as np
import pandas as pd
import csv
```

```python
# distance : 各停留所間の距離行列
# demand : 各停留所の需要
# max_num : 1つの停留所で乗せることができる最大数
distance = pd.read_csv("distance.csv", header=None)
demand = pd.read_csv("demand.csv", header=None)
max_num = 5
distance
```

```python
# div : 分割数
div = demand//max_num
for i in demand:
    if (demand[i][0]%max_num) != 0:
        div[i][0] = div[i][0] + 1
# div_sum : 分割数の和
div_sum = np.sum(div, axis=1)

div2 = []
for i in range(0, len(distance[0])-1):
    div2.append(div[i][0])
div2
```

```python
result = []
result.append([0 for col in range(div_sum[0])])
for i in range(1, len(distance[0])):
    a = [0]
    for j in range(1, len(distance[0])):
        a = a + [distance[i][j]]*div2[j-1]
    for k in range(0, div2[i-1]):
        result.append(a)
result
```

```python
# ファイル書き込み　
with open('data.csv', 'w') as file:
    writer = csv.writer(file, lineterminator='\n')
    writer.writerows(result)
```
